```python
# Create a dictionary to store adjacency matrices for each month
adjacency_matrices = {}

# Loop through each month
for month in range(1, 13):
    # Filter the data for the current month
    data_month = data_2022[data_2022['mes'] == month]
    
    # Create the adjacency matrix for the current month
    adjacency_matrix = data_month.pivot_table(index='co_muni_ori', columns='co_muni_des', values='aero_pass', aggfunc='sum', fill_value=0)
    
    # Save the adjacency matrix for the current month in the dictionary
    adjacency_matrices[month] = adjacency_matrix

# Display or save the adjacency matrices for each month
for month, adjacency_matrix in adjacency_matrices.items():
    print(f"Adjacency matrix for month {month}:\n{adjacency_matrix}\n")
```

```python
import pandas as pd
from multiprocessing import Pool, cpu_count

# Function to process each adjacency matrix
def process_adjacency_matrix(args):
    matrix, lst_value = args
    lis = [0] * len(matrix)
    d = {lst_value: lis}
    dtf = pd.DataFrame(data=d)
    dtf.set_index(matrix.index)
    return dtf

if __name__ == '__main__':
    # Number of processes to run in parallel
    num_processes = cpu_count()

    # Create a pool of processes
    pool = Pool(processes=num_processes)

    # Apply the process_adjacency_matrix function to each adjacency matrix and corresponding lst value in parallel
    results = pool.map(process_adjacency_matrix, zip(adjacency_matrices.values(), lst))

    # Close the pool of processes to free resources
    pool.close()
    pool.join()

    # Create a dictionary to store processed DataFrames for each month
    processed_dfs = {}

    # Join the results to the original adjacency_matrices dictionary
    for ((month, matrix), result_df) in zip(adjacency_matrices.items(), results):
        processed_dfs[month] = matrix.join(result_df)

    # Now processed_dfs contains the processed DataFrame for each month, where each DataFrame is joined with the corresponding adjacency matrix
```
